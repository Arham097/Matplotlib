### **`Problem-1:`** Draw a line plot of which, the x-axis is the "Year" and the y-axis is sum of "PM2.5" of two countries Iran and China. 

```python
df = pd.read_csv('Most Polluted Cities.csv')
iran = df[df['Country'] == 'Iran'].groupby('Year')['PM2.5'].sum().reset_index()
china =  df[df['Country'] == 'China'].groupby('Year')['PM2.5'].sum().reset_index()
plt.plot(iran['Year'], iran['PM2.5'], label='Iran')
plt.plot(china['Year'], china['PM2.5'], label='China')
plt.title('PM2.5 Over the years')
plt.xlabel('Year')
plt.ylabel('PM2.5')
plt.xticks(df['Year'].unique())
plt.xlim(2011,2018)
plt.legend()


# iran

```

### **Problem-1:** Make a subplots which have 2 plots.

- For the first chart, draw a scatter plot "*Monitored Cap.(MW)*" vs "*Total Cap. Under Maintenance (MW)*" of top 5 most frequent power stations. Then draw the lines which indicate the average values of these two columns. Change the colors according to the names of the Power Stations.
- For the second chart, draw a scatter plot "*Monitored Cap. (MW)*" vs "*Actual(MU)*" of the top 5 most frequent power stations. Also draw the lines which indicates the average values of these two columns. Change the colors according to the names of the Power Stations.


```python
top5 = df['Power Station'].value_counts().sort_values(ascending=False).head(5).index
temp = df[df['Power Station'].isin(top5)]
temp

fig, ax = plt.subplots(figsize=(10,10), nrows=2, ncols=1, sharex=True)

for station in top5:
  subset = temp[temp['Power Station']== station]
  ax[0].scatter(
    subset['Monitored Cap.(MW)'],
    subset['Total Cap. Under Maintenace (MW)'],
    label = station
  )


for station in top5:
  subset = temp[temp['Power Station']== station]
  ax[1].scatter(
    subset['Monitored Cap.(MW)'],
    subset['Actual(MU)'],
    label = station
  )



avg_x1 = temp['Monitored Cap.(MW)'].mean()
avg_y1 = temp['Total Cap. Under Maintenace (MW)'].mean()
avg_y2 = temp['Actual(MU)'].mean()

ax[0].axvline(avg_x1, color='gray', linestyle='--', label='Avg. Monitored Cap.(MW)')
ax[0].axhline(avg_y1, color='black', linestyle='--', label='Avg. Cap. Under Maintenance (MW)')
ax[1].axvline(avg_x1, color='gray', linestyle='--', label='Avg. Monitored Cap.(MW)')
ax[1].axhline(avg_y2, color='black', linestyle='--', label='Avg. Actual(MU)')

ax[0].set_xlabel('Monitored Cap.(MW)')
ax[1].set_xlabel('Monitored Cap.(MW)')
ax[0].set_ylabel('Total Cap. Under Maintenance (MW)')
ax[1].set_ylabel('Actual(MU)')
ax[0].set_title('Top 5 Frequent Power Stations: Capacity vs Maintenance')
ax[1].set_title('Top 5 Frequent Power Stations: Capacity vs Actual')
ax[0].legend()
ax[1].legend()
ax[0].grid()
ax[1].grid()

plt.show()
# temp.columns
```

### **Problem-3:** Make a 3D *Surface* plot of this below mathematical equation.

$$z = |x| - |y|$$

```python
import numpy as np
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
# Create a meshgrid of x and y values
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)

# Compute Z = |X| - |Y|
Z = np.abs(X) - np.abs(Y)

p = ax.plot_surface(X,Y,Z, cmap='viridis')
plt.colorbar(p)

```

### **Problem-4:** Draw the 3D *Contour plot* of this below equation:

$$z = |x| - |y|$$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a meshgrid of x and y values
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)

# Compute Z = |X| - |Y|
Z = np.abs(X) - np.abs(Y)

# Create 3D contour plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot( projection='3d')

# Plot contour
contour = ax.contour3D(X, Y, Z, 50, cmap='viridis')

# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z = |X| - |Y|')
ax.set_title('3D Contour Plot of z = |x| - |y|')

plt.show()
```

### **Problem-5:** Draw a second type of Countour plot of the below equation:

$$z = |x| - |y|$$

```python
# Create a meshgrid of x and y values
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)

# Compute Z = |X| - |Y|
Z = np.abs(X) - np.abs(Y)

# Create 3D contour plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot()

# Plot contour
contour = ax.contourf(X, Y, Z, 50, cmap='viridis')
plt.colorbar(contour)
# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('3D Contour Plot of z = |x| - |y|')

plt.show()
```

### `Problem-6` Use Pandas plot functions

* Line plot of closing value of top 5 Stocks in Year 2020.
* Take top 5 stocks based on total turnover in Year 2020

```python
df['Date'] = pd.to_datetime(df['Date'])
only_2020 = df[df['Date'].dt.year==2020]
top5_1 = only_2020.groupby('Symbol').agg(
  {
    'Close': 'max'
  }
).sort_values('Close',ascending=False).head(5)

top5_2 = only_2020.groupby('Symbol').agg(
  {
    'Turnover': 'max'
  }
).sort_values('Turnover',ascending=False).head(5)

top5_1.plot(ylabel='Closing', title="Stock's Closing Price")
top5_2.plot(ylabel='Turnover', title="Stock's Highest turnover")
```
