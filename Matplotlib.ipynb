# **Matplotlib**

#### ðŸ“Š Matplotlib is a popular Python library used for creating visualizations like line plots, bar charts, scatter plots, histograms, and more.
#### It gives you full control over every part of a plot â€” titles, labels, colors, legends, axes, and more â€” making it great for both basic and advanced graphing.

#### Itâ€™s especially useful in data analysis and scientific computing, and works well with libraries like NumPy and Pandas.

### **Types of Data**

- Numerical Data
- Categorical Data

```python
# import the library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('default')
```

### **2D Line plot**

- Bivariate Analysis
- categorical -> numerical and numerical -> numerical
- Use case - Time series data

```python
# plotting a simple function
price = [48000,54000,57000,49000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]
plt.plot(year,price)
```

```python
# from a pandas dataframe
batsman = pd.read_csv('sharma-kohli.csv')
batsman

plt.plot(batsman['index'], batsman['V Kohli'])
```

### **Plotting multiple plots**

```python

plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

### **Labels, Title**

```python
# labels title
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

### **colors(hex) and line(width and style) and marker(size)**

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='#D91345', linestyle='dashed', linewidth='2', marker='o')
plt.plot(batsman['index'], batsman['RG Sharma'], color='#FC00D6', linestyle='dotted', linewidth='3', marker='*', markersize=10)

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

### **legend -> location**

```python
# legend -> location
plt.plot(batsman['index'],batsman['V Kohli'],color='#D9F10F',linestyle='solid',linewidth=3,marker='D',markersize=10,label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#FC00D6',linestyle='dashdot',linewidth=2,marker='o',label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend() # choose best position on figure where space available
plt.legend(loc='upper right') # can change to desired location through (loc) parameter
```

### **limiting axes**

#### - To handle outliers that effect data visualization

```python
price = [48000,54000,57000,49000,47000,45000,4500000]  # the outlier 4500000 cause remaining other values flat(straight) to bottom on visualization
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year, price)
```

```python
price = [48000,54000,57000,49000,47000,45000,4500000]  # the outlier 4500000 cause remaining other values flat(straight) to bottom on visualization
year = [2015,2016,2017,2018,2019,2020,2021]

plt.plot(year, price)
plt.ylim(0,75000)
plt.xlim(2017,2021)
```

```python
# color and label
df = pd.read_csv('batsman_season_record.csv')

# for side by side bars of different years there is no built in function so we have to make it with custom logic
plt.bar(np.arange(df.shape[0]) - 0.2,df['2015'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0]) ,df['2016'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]) + 0.2,df['2017'],width=0.2,color='blue')

plt.xticks(np.arange(df.shape[0]),df['batsman'])

plt.show()
```

```python
# a problem => if there are x axis values are so long so it overlaps
children = [10,20,40,10,30]
colors = ['red red red red red red','blue blue blue blue','green green green green green','yellow yellow yellow yellow ','pink pinkpinkpink']

plt.bar(colors,children,color='black')
plt.xticks(rotation='vertical') # makes x values vertical so it shows smoothly without overlap
```

### **Stacked Bar chart**

- Gives total + individual values

```python
plt.bar(df['batsman'],df['2015'], label='2015')
plt.bar(df['batsman'],df['2016'], bottom=df['2015'] , label='2016')
plt.bar(df['batsman'],df['2017'], bottom=(df['2015']+df['2016']) , label='2017')

plt.legend()
```

## **Annotations**
- name (values) at the dot

```python
batters = pd.read_csv('batter.csv')
batters.shape

sample_df = batters.head(100).sample(25,random_state=5)
sample_df
```

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'], sample_df['strike_rate'], s=sample_df['runs'])

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

## **Horizontal and Vertical lines**

For Analysis in particular Area

```python
plt.figure(figsize=(18,10))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
plt.xlabel('Avg')
plt.ylabel("SR")
plt.title("Avg vs SR Comparison")

plt.axhline(130,color='red')
plt.axvline(30,color='red')

for i in range(sample_df.shape[0]):
  plt.text(sample_df['avg'].values[i], sample_df['strike_rate'].values[i], sample_df['batter'].values[i])
```

## **Subplots**

```python
# This is first way to make graphs
plt.figure(figsize=(15,6))
plt.scatter(batters['avg'],batters['strike_rate'])
plt.title('Something')
plt.xlabel('Avg')
plt.ylabel('Strike Rate')

plt.show()
```

```python
# This is another way to make graphs

fig, ax = plt.subplots(figsize=(15,6))
ax.scatter(batters['avg'],batters['strike_rate'], marker='+')
ax.set_title('Avg vs SR Comparison')
ax.set_xlabel('Avg')
ax.set_ylabel('SR')

fig.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(12,8))
ax[0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[1].scatter(batters['avg'],batters['runs'])

ax[0].set_title('Avg Vs Strike Rate')
ax[0].set_ylabel('Strike Rate')


ax[1].set_title('Avg Vs Runs')
ax[1].set_ylabel('Runs')
ax[1].set_xlabel('Avg')
```

```python
fig, ax = plt.subplots(nrows=2,ncols=2,figsize=(14,10))
ax[0][0].scatter(batters['avg'],batters['strike_rate'],color='red')
ax[0][1].scatter(batters['avg'],batters['runs'])

ax[0][0].set_title('Avg Vs Strike Rate')
ax[0][0].set_ylabel('Strike Rate')
ax[0][0].set_xlabel('Avg')


ax[0][1].set_title('Avg Vs Runs')
ax[0][1].set_ylabel('Runs')
ax[0][1].set_xlabel('Avg')

df = pd.read_csv('batsman_season_record.csv')
ax[1][0].bar(df['batsman'],df['2015'])
ax[1][0].set_title('Batsman Runs in 2015 ')
ax[1][0].set_xlabel('Batsman')
ax[1][0].set_ylabel('Runs Scored')

df2 = pd.read_csv('gayle-175.csv')
ax[1][1].pie(df2['batsman_runs'],labels=df2['batsman'], autopct = "%0.1f%%")
ax[1][1].set_title('Match Contributions ')
```

```python
# Another way to make subplots
fig = plt.figure()
ax1 = fig.add_subplot(2,1,1)  # (2,1,1) means there will be 2 graph, 1 shows 1st column , 1 shows 1st graph/row
ax1.hist(batters['runs'])

ax2 = fig.add_subplot(2,1,2) # (2,1,2) means there will be 2 graph, 1 shows 1st column , 2 shows 1st graph/row
ax2.hist(batters['avg'])
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1, sharex=True ) # both graph share same x labels
ax[0].scatter(batters['avg'],batters['strike_rate'])
ax[0].set_xlabel('Avg')
ax[0].set_ylabel('SR')
ax[1].scatter(batters['avg'], batters['runs'])
ax[1].set_xlabel('Avg')
ax[1].set_ylabel('Runs Scored')
```

### 3D Line Plot

```python
x = [0,1,5,25]
y = [0,10,13,0]
z = [0,13,20,9]

fig = plt.figure()

ax = plt.subplot(projection='3d')

ax.scatter3D(x,y,z, s=[100,100,100,100])
ax.plot3D(x,y,z, color='red')
```

### 3D Surface Plots

To Visualize loss function

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)

xx,yy = np.meshgrid(x,y)
z = xx**2 + yy**2
# print(z)
p = ax.plot_surface(xx,yy,z, cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)

fig = plt.figure(figsize=(12,8))

ax = plt.subplot(projection='3d')

p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

### Contour Plots

To visualize 3d plots into 2d

```python
fig = plt.figure(figsize=(12,8))

ax = plt.subplot()

p = ax.contourf(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

```python
z = np.sin(xx) + np.cos(yy)
fig = plt.figure()
ax = fig.add_subplot()
p = ax.contourf(xx,yy,z)
fig.colorbar(p)
```
